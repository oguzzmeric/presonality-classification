```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

```

```python
data = pd.read_csv("train.csv")

data.head()
```

```python
data.isnull().sum()
```

```python
data.groupby("Personality (Class label)").mean()
```

```python
data = data.replace({"Personality (Class label)":{"dependable":1,"extraverted":2,"lively":3,"responsible":4,"serious":5}})
data = data.replace({"Gender":{"Male":0,"Female":1}})
```

```python
data
```

"dependable":1,"extraverted":2,"lively":3,"responsible":4,"serious":5

```python
x = data.drop("Personality (Class label)",axis=1)
y = data["Personality (Class label)"]
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)


```

```python
predictiontree = dtree.predict(x_train)
from sklearn.metrics import accuracy_score
acc_scr = accuracy_score(y_train,predictiontree)
print(acc_scr)
```
